Given a deck of poker cards (total 52 card names in a list in random order). You need to
 
(1) Define a function which will draw a card from the top of deck.  
  In the function, convert card names to points:

    - Ace:    11 points
    - Jack:   10 points
    - Queen:  10 points
    - King:   10 points
    - Others: points = card face number

    Return the card's points and card name.

```python
# Hint: Use the below list for deck of cards
import random 

values = [ "ace", "2", "3", "4", "5", "6", "7","8", "9", "10", "jack", "queen", "king" ]
suits  = [ "clubs", "diamonds", "hearts", "spades" ]
deck = [ v +"_of_"+ s for s in suits for v in values  ] 

random.shuffle(deck) 
print(deck)
```

```python
import ipywidgets as widgets
from IPython.display import Image, HTML, display


def show(*card_names):
    length = len(card_names)
    if length == 0:
        print(" There is no card!")
    else:
        card_list = []
        for i in range(length):
            path_name = './deck/'+ card_names[i] + '.png'
            img = open(path_name, 'rb').read()
            img_resize = widgets.Image(value= img, format='png', width=60, height=40)
            card_list.append(img_resize)
        sidebyside = widgets.HBox(card_list)
        display(sidebyside)

# for one card
show("queen_of_clubs")

# for two card
show("jack_of_clubs", "ace_of_diamonds")

# for three card
show("10_of_hearts", "jack_of_clubs", "ace_of_diamonds")
```
